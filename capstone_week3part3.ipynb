```python
url='https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'
page = urlopen(url).read().decode('utf-8')
soup = BeautifulSoup(page,'html.parser')
table=soup.body.table.tbody
```

```python
def cell(row_element):
    cells = row_element.find_all('td')
    row=[]
    
    for cell in cells:
        if cell.a:
            if cell.a.text:              
                row.append(cell.a.text)
                continue
        row.append(cell.string.strip())
#     print(row)
    return row
```

```python
def row_data():
    data = []
    for row in table.find_all('tr'):
        data_row = cell(row)
        if len(data_row)!=3:
            continue
#         print(data_row)
        data.append(data_row)
    return data

data= row_data()
# print(data)
columns=['Postcode','Borough','Neighbourhood']
df = pd.DataFrame(data,columns=columns)
df
```

```python
df_cleanborough=df[df.Borough != 'Not assigned']
df1 = df_cleanborough.sort_values(by=['Postcode','Borough'])
df1.reset_index(inplace=True)
df1.drop('index',axis=1,inplace=True)
# print(df1)

df_post= df1['Postcode']
df_post.drop_duplicates(inplace=True)
df2=pd.DataFrame(df_post)
df2['Borough']=''
df2['Neighbourhood']=''
df2.reset_index(inplace=True)
df2.drop('index',axis=1,inplace=True)
df1.reset_index(inplace=True)
df1.drop('index',axis=1,inplace=True)



# print(df2)
for i in df2.index:
    for j in df1.index:
        if df1.iloc[j,0]==df2.iloc[i,0]:
            df2.iloc[i,1]=df1.iloc[j,1]
            df2.iloc[i,2]= df2.iloc[i,2]+','+df1.iloc[j,2]
# print(df2)
for i in df2.index:
    s=df2.iloc[i,2]
    
    if s[0]==',':
        s=s[1:]
        df2.iloc[i,2]=s
df2
# df2
# df1.drop(index)
# print(df1)
```

```python
toronto_map = folium.Map(location=[43.65, -79.4], zoom_start=12)

X = TorontoData['Latitude']
Y = TorontoData['Longitude']
Z = np.stack((X, Y), axis=1)

kmeans = KMeans(n_clusters=4, random_state=0).fit(Z)

clusters = kmeans.labels_
colors = ['red', 'green', 'blue', 'yellow']
TorontoData['Cluster'] = clusters

for latitude, longitude, borough, cluster in zip(TorontoData['Latitude'], TorontoData['Longitude'], TorontoData['Borough'], TorontoData['Cluster']):
    label = folium.Popup(borough, parse_html=True)
    folium.CircleMarker(
        [latitude, longitude],
        radius=5,
        popup=label,
        color='Yellow',
        fill=True,
        fill_color=colors[cluster],
        fill_opacity=0.7).add_to(toronto_map)  

toronto_map
```
